```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
sns.load_dataset
```

```python
tips = sns.load_dataset('tips')
```

```python
sns.distplot(tips['total_bill'])
```

```python
sns.distplot(tips['total_bill'], kde=False) # 파란 선 보기 싫으면 kde 설정!
```

```python
sns.distplot(tips['total_bill'], hist=False) # 막대 보기 싫으면 hist 설정!
plt.title("Total Bill Density")
plt.xlabel("Total Bill")
plt.ylabel("Unit Probability")
plt.show()
```

```python
sns.distplot(tips['total_bill'], rug=True)
```

```python
sns.countplot('day', data=tips)
```

```python
sns.regplot(x='total_bill', y='tip', data=tips, fit_reg=False)
```

```python
g = sns.jointplot(x="total_bill", y= 'tip', data=tips, alpha=0.5)
g.set_axis_labels(xlabel = 'Totall Bill', ylabel = 'Tip')
g.fig.suptitle("Joint Plot of Total Bill", fontsize=10, y= 1.03)
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips, kind='hex')
```

```python
sns.kdeplot(data=tips['total_bill'], data2=tips['tip'], shade=True)
```

```python
sns.barplot(x='time', y= 'total_bill', data=tips)
```

```python
sns.boxplot(x='time', y='total_bill', data=tips)
```

```python
sns.violinplot(x='time', y='total_bill', data=tips)
```

```python
sns.pairplot(tips)
```

```python
sns.pairplot(tips, diag_kind='kde')
```

```python
g = sns.PairGrid(tips)
g = g.map_upper(sns.regplot)  # 행렬 LU분해 생각해봐! 
g = g.map_lower(sns.kdeplot)
g = g.map_diag(sns.distplot, rug=True)
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.boxplot(x='day', y='total_bill', data=tips)
plt.show()
```

```python
sns.set_style('whitegrid')
sns.boxplot(x='day', y='total_bill', hue="smoker", data=tips, palette='Set3')
```

```python
sns.boxplot(x='day', y='total_bill', data=tips)
sns.swarmplot(x='day', y='total_bill', data=tips, color='.5')
```

```python
g = sns.FacetGrid(tips, col='time')
g.map(sns.distplot, 'total_bill', rug=True)
```

```python
g = sns.FacetGrid(tips, col='day', hue='sex')
g = g.map(plt.scatter, 'total_bill', 'tip')
g = g.add_legend()
```

```python
g = sns.FacetGrid(tips, col="time", row="smoker", hue="sex")
g.map(plt.scatter, 'total_bill', 'tip')
```
