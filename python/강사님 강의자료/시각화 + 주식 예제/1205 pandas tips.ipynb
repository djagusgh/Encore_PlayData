```python
df = sns.load_dataset('iris')
df.head()
```

```python
sum_by_species = df.groupby('species').sum()
```

```python
sum_by_species
```

```python
var = 'sepal_width'
```

```python
sum_by_species[var].plot(kind="pie")
plt.title("꽃받침의 너비로 분류한 붓꽃")
plt.ylabel('')
plt.show()
```

```python
title = "종에 따른 전체 측정값 분류"
sum_by_species.plot(kind="pie", subplots=True, layout=(2,2), legend=False, 
                    figsize=(10,8), title=title)
plt.show()
```

```python
sum_by_species[var].plot(kind='bar')
```

```python
plt.figure(figsize=(10,8))
print(df['species'].unique())
for spec in df['species'].unique():
    forspec = df[df['species'] == spec]
    forspec['petal_length'].plot(kind='hist', alpha=0.4, label = spec)
plt.legend(loc = 'upper right')
plt.suptitle('종에 따른 꽃잎 길이')
plt.show()
```

```python
col = df['petal_length']
average = col.mean()
print("average ==========", average)
median = col.quantile(0.5)
print("median ==========", median)
percenttile25 = col.quantile(0.25)
print("percenttile25 ==========", percenttile25)
percenttile75 = col.quantile(0.75)
print("percenttile75 ==========", percenttile75)

# outlier를 제거한 후 평균
clean_avg = col[(col>percenttile25) & (col<percenttile75)].mean()
print("clean_avg ==========", clean_avg)
```
