```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
type(tips)
```

```python
fig = plt.figure()
axes1 = fig.add_subplot(1,1,1)
```

```python
axes1.hist(tips['total_bill'], bins = 10) #10개 구간으로 나눔
```

```python
axes1.set_title('HIstogram of Total Bill')
axes1.set_xlabel('Frequency')
axes1.set_ylabel('Total Bill')
fig
```

```python
scatter_plot = plt.figure()
axes1 = scatter_plot.add_subplot(1,1,1)
color_type = []
for i in tips['smoker']:
    if i == 'No':
        color_type.append('r')
    else:
        color_type.append('black')
axes1.scatter(tips['total_bill'], tips['tip'], c = color_type)
axes1.set_title('Scatterplot of Total Bill vs Tip')
axes1.set_xlabel('Total Bill')
axes1.set_ylabel('Tip')
```

```python
boxplot = plt.figure()
axes1 = boxplot.add_subplot(1,1,1)

females_tip = tips[tips['sex'] == 'Female']['tip']
males_tip = tips[tips['sex'] == 'Male']['tip']
axes1.boxplot([females_tip, males_tip], labels = ['Female', 'Male'])

'''
axes1.boxplot(
    [tips[tips['sex'] == 'Female']['tip'],
    tips[tips['sex'] == 'Male']['tip']],
    labels = ['Female', 'Male'])
'''

axes1.set_title('Boxpplot of Tips by Sex')
axes1.set_xlabel('Sex')
axes1.set_ylabel('Tip')
```

```python
#다변량 데이터 그래프
```

```python
def recode_sex(sex):
    if sex == 'Female':
        return 0
    else:
        return 1
```

```python
tips['sex_color'] = tips['sex'].apply(recode_sex) # apply 고차함수 데이터.apply(적용시킬 함수)
```

```python
tips.head()
```

```python
scatter_plot = plt.figure()
axes1= scatter_plot.add_subplot(1,1,1,)
axes1.scatter(
    x = tips['total_bill'],
    y = tips['tip'],
    s = tips['size'] * 15,
    c = tips['sex_color'],
    alpha = 0.5 #투명도
)

axes1.set_title('Total Bill  vs Tip Colored by Sex and Sized by Size')
axes1.set_xlabel('Total Bill')
axes2.set_ylabel('Tip')
```
