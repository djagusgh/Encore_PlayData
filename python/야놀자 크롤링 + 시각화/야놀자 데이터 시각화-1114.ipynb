```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from bs4 import BeautifulSoup
```

```python
def make_location(address): # ex) 서울특별시 광진구 화양동 24-15 H Avenue 건대점 -> 서울특별시 광진구
    temp = address.split(" ")
    loc = temp[0] + " " + temp[1]
    return loc
```
