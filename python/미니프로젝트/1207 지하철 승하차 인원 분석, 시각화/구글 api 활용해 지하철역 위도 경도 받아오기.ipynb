```python
# subway_2017내 지하철의 위도, 경도 리스트 만들기

lat_list = [] # 위도 리스트
lng_list = [] # 경도 리스트
for each in sub_list:
    try:
        tmpMap = gmaps.geocode(each, language="ko")
        # 위도 경도를 받고
        lat = tmpMap[0]['geometry']['location']['lat']
        lng = tmpMap[0]['geometry']['location']['lng']
    except:
        try:
            # tmpMap이 []로 받아지는 경우 존재 -> lat, lng받을 때 indexError 발생 , 이유를 찾아보니까
            # ex) '옥수' 로 geocode에 넣으면 빈 리스트 출력, '옥수역'으로 geocode에 넣어야 원하는 값이 나옴
            each += "역" # 역을 붙이면 다 정보 존재함
            tmpMap = gmaps.geocode(each, language="ko")
            # 위도 경도를 받고
            lat = tmpMap[0]['geometry']['location']['lat']
            lng = tmpMap[0]['geometry']['location']['lng']
        except:
            # 여기까지 오는 예외 케이스;; : (우이실선설 삼양 -> 삼양도 없고 삼양역도 없다.)
            lat = None
            lng = None
            print(" 뭐야 이건", each)
        
    # 위도, 경도 리스트에 값 저장
    lat_list.append(lat)
    lng_list.append(lng)
```

```python
# 길이 다 같은지 체크
print("지하철 갯수는 " , len(sub_list))
print("위도 갯수는 " ,len(lat_list))
print("경도 갯수는 " ,len(lng_list))
```

```python
# 세 리스트를 DataFrame 에 저장
sub_position = pd.DataFrame({'SUB_STA_NM' : sub_list,
                            'latitude' : lat_list,
                            'longitude' : lng_list})
```

```python
# csv로 저장
sub_position.to_csv("sub_position.csv")
```
