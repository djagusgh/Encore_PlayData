## differ : 탑승객 수 - 하차인원 수 (파생변수 생성)

```python
# differ : 탑승객 수 - 하차인원 수 (파생변수 생성)
tmpdf = subway_2017[use_columns]
tmpdf["SEVEN_differ"] = tmpdf["SEVEN_RIDE_NUM"] - tmpdf["SEVEN_ALIGHT_NUM"]
tmpdf["EIGHT_differ"] = tmpdf["EIGHT_RIDE_NUM"] - tmpdf["EIGHT_ALIGHT_NUM"]
tmpdf["EIGHTEEN_differ"] = tmpdf["EIGHTEEN_RIDE_NUM"] - tmpdf["EIGHTEEN_ALIGHT_NUM"]
tmpdf["NINETEEN_differ"] = tmpdf["NINETEEN_RIDE_NUM"] - tmpdf["NINETEEN_ALIGHT_NUM"]
tmpdf["TWENTY_differ"] = tmpdf["TWENTY_RIDE_NUM"] - tmpdf["TWENTY_ALIGHT_NUM"]
```

```python
# 1년치 -> 월 별로 DataFrame 쪼개기

tmp01 = tmpdf[tmpdf["USE_MON"] == 201701]
tmp02 = tmpdf[tmpdf["USE_MON"] == 201702]
tmp03 = tmpdf[tmpdf["USE_MON"] == 201703] 
tmp04 = tmpdf[tmpdf["USE_MON"] == 201704]
tmp05 = tmpdf[tmpdf["USE_MON"] == 201705] 
tmp06 = tmpdf[tmpdf["USE_MON"] == 201706] 
tmp07 = tmpdf[tmpdf["USE_MON"] == 201707] 
tmp08 = tmpdf[tmpdf["USE_MON"] == 201708] 
tmp09 = tmpdf[tmpdf["USE_MON"] == 201709] 
tmp10 = tmpdf[tmpdf["USE_MON"] == 201710] 
tmp11 = tmpdf[tmpdf["USE_MON"] == 201711]
tmp12 = tmpdf[tmpdf["USE_MON"] == 201712] 
```

## year_pivot : 2017년 역별 출퇴근 시간의 탑승  - 하차 인원 pivot
#### (1월 ~12월 데이터 평균 냄)

```python
tmp_list = [tmp01, tmp02, tmp03, tmp04, tmp05, tmp06, tmp07, tmp08, tmp09, tmp10, tmp11, tmp12]

tmp01 = tmp01.groupby(['SUB_STA_NM'], as_index=False).sum()
year_pivot = pd.pivot_table(tmp01, index='SUB_STA_NM')
for i, each in enumerate(tmp_list):
    if i == 0:
        continue
    year_pivot += pd.pivot_table(each.groupby(['SUB_STA_NM'], as_index=False).sum(), index='SUB_STA_NM', aggfunc = np.mean)

year_pivot = year_pivot / 12
```

```python
year_pivot.shape
```

```python
# pivot 을 DataFrame으로!
final = pd.DataFrame({
    'SUB_STA_NM' : year_pivot.index,
    'SEVEN_differ' : year_pivot['SEVEN_differ'].values,
    'EIGHT_differ' : year_pivot['EIGHT_differ'].values,
    'EIGHTEEN_differ' : year_pivot['EIGHTEEN_differ'].values,
    'NINETEEN_differ' : year_pivot['NINETEEN_differ'].values,
    'TWENTY_differ' : year_pivot['TWENTY_differ'].values
})
```

# 지하철 승하차 인원, 위 경도 데이터 합치기

```python
# 위도, 경도 데이터 합치기!
final = pd.merge(final, sub_pos, on ='SUB_STA_NM')
```

# min-max 정규화

```python
# min - max 정규화 (0 ~ 100)
def reRange(x, oldMin, oldMax, newMin, newMax):
    return (x - oldMin)*(newMax - newMin) / (oldMax - oldMin) + newMin

final['SEVEN_differ'] = reRange(final['SEVEN_differ'], min(final['SEVEN_differ']), max(final['SEVEN_differ']),1 , 100)
final['EIGHT_differ'] = reRange(final['EIGHT_differ'], min(final['EIGHT_differ']), max(final['EIGHT_differ']),1 , 100)
final['EIGHTEEN_differ'] = reRange(final['EIGHTEEN_differ'], min(final['EIGHTEEN_differ']), max(final['EIGHTEEN_differ']),1 , 100)
final['NINETEEN_differ'] = reRange(final['NINETEEN_differ'], min(final['NINETEEN_differ']), max(final['NINETEEN_differ']),1 , 100)
final['TWENTY_differ'] = reRange(final['TWENTY_differ'], min(final['TWENTY_differ']), max(final['TWENTY_differ']),1 , 100)
```

```python
# 중복값이 왜 있는지는 모르겠는데 중복값 제거!
final = final.drop_duplicates()
```

# 7시 ~ 8시 평균

```python
# 7~8시 기준  상위 5개, 하위 5개 뽑아내기
final7_max5 = final.sort_values("SEVEN_differ", ascending = False)[:10]
final7_min5 = final.sort_values("SEVEN_differ", ascending = True)[:10]

# 상위 5개 역 이름
sub_sta_nm_max =list(final7_max5['SUB_STA_NM'])
# 하위 5개 역 이름
sub_sta_nm_min =list(final7_min5['SUB_STA_NM'])


# 상위 5개 역 위, 경도
lat_max5 = list(final7_max5['latitude'])
lng_max5 = list(final7_max5['longitude'])
lat_lng_max5 = [[i, j] for i, j in zip(lat_max5, lng_max5)]
# 하위 5개 역 위, 경도
lat_min5 = list(final7_min5['latitude'])
lng_min5 = list(final7_min5['longitude'])
lat_lng_min5 = [[i, j] for i, j in zip(lat_min5,lng_min5)]

map1 = folium.Map(location=[37.5502, 126.982], zoom_start=11)

for n in final.index:
    
    # red 일수록 탑승인원 > 하차 인원
    if final['SEVEN_differ'][n] >= 50:
        folium.CircleMarker([final['latitude'][n], final['longitude'][n]], radius = final['SEVEN_differ'][n] * 0.15,
                           color="red", fill=True, fill_color='red').add_to(map1)
    else:
    # blue 일수록 하차 인원 > 탑승 인원, radius : 100에서 뺀 값을 넣어야 함! min에 가까워질 수록
    # 파란 원이 커져야 하는데, 그대로 넣으면 min일수록 파란 원이 작게 그려짐
        folium.CircleMarker([final['latitude'][n], final['longitude'][n]], radius = (100 - final['SEVEN_differ'][n]) * 0.15,
                           color="#3186cc", fill=True, fill_color='#3186cc').add_to(map1)
        
# max 5개 값 빨간 info-icon으로 추가!
for loc, name in zip(lat_lng_max5, sub_sta_nm_max):
    folium.Marker(loc, popup= name ,icon = folium.Icon(color ='red', icon='info-sign')).add_to(map1)
for loc, name in zip(lat_lng_min5, sub_sta_nm_min):
    folium.Marker(loc, popup=name ,icon = folium.Icon(color ='blue', icon='info-sign')).add_to(map1)
        
map1
```
