```python
from itertools import permutations

def solution(baseball):
    answer = 0
    # 1) 1 ... 9 로 만들 수 있는 모든 세자리 정수의 조합 담는 리스트(allList) 만들기
    allList = []
    temp = permutations(list(range(1, 10)), 3)
    for each in temp:
        threeDigits = ""
        for each_num in each:
            threeDigits += str(each_num)
        allList.append(int(threeDigits))
        
    
    for each_num in allList:
        count = 0 # baseball 내 각 숫자와 Strike, Ball 정보 같을 때마다 + 1
        for guess in baseball:
            guess_num = guess[0]
            strike = 0
            ball = 0
            each_num = str(each_num); guess_num = str(guess_num) # 리스트 처럼 활용하기 위해..!
            # 2) guess_num과 each_num의 strike,ball 구하기
            for idx, val in enumerate(each_num):
                if each_num[idx] == guess_num[idx]:
                    strike += 1
                elif val in guess_num:
                    ball += 1
            
            if (strike, ball) == (guess[1], guess[2]):
                count += 1
        
        if count == len(baseball): # 3) baseball 내의 모든 숫자와 정보가 같으면 answer + 1
            answer += 1    
    return answer
```

```python
solution([[123, 1, 1], [356, 1, 0], [327, 2, 0], [489, 0, 1]])
```
